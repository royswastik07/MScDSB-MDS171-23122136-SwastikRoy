```python
# create a Stack class
class Stack:
    def __init__(self):   # create a constructor 
        self.stack=[]   #define a empty list 
    def pushitem(self,item): # create a function for push the items in the list / stack
        self.stack.append(item)
    def popitem(self):   # create a function for pop the items in the list / stack
        return self.stack.pop()
    def viewstack(self):  # create a function for see all the items from stack
        for item in self.stack:
            print(item)
    def size(self):  # create a function for get the size the stack 
        print(len(self.stack))
    def topstack(self):  # create a function fro peek 
        print(self.stack[-1])
    def isEmpty(self):  # create a function to check the stack is empty or not 
        if self.stack==[]:
            print("The Stack Is empty")
        else:
            print("The Stack is Not empty")
    def Menu (self):    # create a menu function 
        while True:
            print("Menu Options")
            print("************************************")
            print("1 PushItem")
            print("2 PopItem")
            print("3 View Full Stack")
            print("4 Size of the stack")
            print("5 View the top Element of the stack")
            print("6 to check the stack is empty or not")
            print("7 Exit ")
            print("************************************")
            choice=int(input("Enter your choice: "))
            if choice== 1:
                item=int(input("Enter: "))
                self.pushitem(item)
            elif choice== 2:
                self.popitem()
            elif choice== 3:
                self.viewstack()
            elif choice== 4:
                self.size()
            elif choice== 5:
                self.topstack()
            elif choice== 6:
                self.isEmpty()
            elif choice== 7:
                print("Thank You")
                break
            else:
                print("Chose a Valid Output")
```

```python
# create a object
obj=Stack()   
```

```python
# called the menu function
obj.Menu()
```
