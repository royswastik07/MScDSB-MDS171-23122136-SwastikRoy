```python
def GSMS(name,phno,date,item,quantity,price,total_bill):
    with open("FarmersHarvest.txt","a+") as f:
        f.write(f"name:\t{name}\n")
        f.write(f"phno:\t{phno}\n")
        f.write(f"item:\t{item}\n")
        f.write(f"Day:\t{date}\n")
        f.write(f"quantity:\t{quantity}\n")
        f.write(f"Price:\t{price}\n")
        f.write(f"Total Bill:\t{total_bill}\n")
    return True
name=input("enter your name: ")
ph_no=input("enter your phone number: ")
item=input("enter items: ")
date=input("enter day: ")
quantity=int(input("enter the quantity: "))
price=int(input("enter the price: "))
total_bill= quantity * price
rest=GSMS(name,ph_no,item,date,quantity,price,total_bill)
if rest:
    print("Orders received!!!!")
    print("-"*30)
    file = open("FarmersHarvest.txt", "r+")
    file_lines = file.readlines()
    for line in file_lines:
        print(file_lines)
    print("-"*30)
else:
    print("Error!!!!")
#analysis
salescount = 0
with open("FarmersHarvest.txt" , "r+") as file :
    for line in file:
        if "Rate" in line:
            Rate = int(line.split("=")[1].strip())
            salescount += Rate
print("The total amount obtained in this day = " , salescount)
```
